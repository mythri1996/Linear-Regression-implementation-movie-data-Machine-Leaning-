LR using Sklearn



```
from sklearn.linear_model import LinearRegression
```

```
x=movie_data.iloc[:,1:].values # taking x values (converting as array)
y=movie_data.iloc[:,0:1].values
```

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=5,test_size=0.20)
```

```
from sklearn.linear_model import LinearRegression
Regression= LinearRegression()
Regression.fit(x_train,y_train)
```

```
#predicting values
y_pred=Regression.predict(x_test)
y_pred
```

```
from sklearn.metrics import r2_score #r2_score? will help to know with documentation
r2_score(y_pred,y_test)
```
